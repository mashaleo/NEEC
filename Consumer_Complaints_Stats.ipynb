```python
import pandas as pd
import numpy as np
from google.colab import files
from datetime import datetime

# Upload the CSV file
uploaded = files.upload()

# Read the CSV file
df = pd.read_csv(next(iter(uploaded)))

# Check if issues are nested within products
issue_product_nested = df.groupby('Product')['Issue'].nunique() == df['Issue'].nunique()

print("Are issues nested within products?")
print(issue_product_nested)

# Convert date columns to datetime
df['Date received'] = pd.to_datetime(df['Date received'])
df['Date sent to company'] = pd.to_datetime(df['Date sent to company'])

# Calculate business days
df['Business days'] = (df['Date sent to company'] - df['Date received']).dt.days

# Calculate statistics of business days by company public response
response_stats = df.groupby('Company public response')['Business days'].agg(['count', 'mean', 'std', 'min', 'max'])

print("\nStatistics of business days by company public response:")
print(response_stats)

# Calculate statistics of business days by issues
issue_stats = df.groupby('Issue')['Business days'].agg(['count', 'mean', 'std', 'min', 'max'])

print("\nStatistics of business days by issues:")
print(issue_stats)
```
